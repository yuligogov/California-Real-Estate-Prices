```python
# CODE/BUSINESS LOGIC

import matplotlib.pyplot as plt
import locale
locale.setlocale( locale.LC_ALL, '' )
import math

# Geography Class Graphs Data from County and State Objects
class Geography():

    # Displays Graph with Inputed Names and X-Axis and Y-Axis Values
    def show_graph(self, start_year, end_year, x_list, y_list, y_dollar_format, graph_title, graph_y_title):
        # X-Axis Labels
        x_years_hightest = x_list[-1]
        x_years_lowest = x_list[0]

        x_years_values = []
        if len(x_list) > 5:
            for i in range(x_years_lowest, x_years_hightest + 1, int((x_years_hightest - x_years_lowest) / 5)):
                x_years_values.append(i)
        else:
            for i in range(x_years_lowest, x_years_hightest + 1):
                x_years_values.append(i)

        custom_ticks_x = x_years_values
        custom_labels_x = x_years_values
        plt.xticks(custom_ticks_x, custom_labels_x)

        # Y-Axis Labels
        y_data_highest = int(y_list[-1])
        y_data_lowest = int(y_list[0])
        for data in y_list:
            if data > y_data_highest:
                y_data_highest = int(data)

        y_values = []
        for i in range(y_data_lowest, y_data_highest, int((y_data_highest - y_data_lowest) / 4)):
            y_values.append(i + 1)

        y_labels = []
        for data in y_values:
            if y_dollar_format:
                y_labels.append(locale.currency(data * 1000, grouping = True))
            else:
                y_labels.append("{:,}".format(round(data * 1000, -3)))

        custom_ticks_y = y_values
        custom_labels_y = y_labels
        plt.yticks(custom_ticks_y, custom_labels_y)

        # Init Graph
        plt.plot(x_list, y_list)

        plt.title(graph_title + self.name + " (" + str(start_year) + "-" + str(end_year) + ")")
        plt.xlabel("Years")
        plt.ylabel(graph_y_title)

        plt.show()


    # Calls show_graph() with House Price Index Data
    def show_hpi_graph(self, start_year, end_year):
        x_years_list = []
        y_hpi_list = self.get_hpi(start_year, end_year)
        for i in range(start_year, end_year + 1):
            x_years_list.append(i)

        self.show_graph(start_year, end_year, x_years_list, y_hpi_list, False, "House Price Index for ", "House Prices")


    # Calls show_graph() with Personal Income Data
    def show_pi_graph(self, start_year, end_year):
        x_years_list = []
        y_pi_list = self.get_pi(start_year, end_year)
        for i in range(start_year, end_year + 1):
            x_years_list.append(i)

        self.show_graph(start_year, end_year, x_years_list, y_pi_list, False, "Personal Income for ", "Thousands of Dollars")


    # Calls show_graph() with Residential Population Data
    def show_rp_graph(self, start_year, end_year):
        x_years_list = []
        y_rp_list = self.get_rp(start_year, end_year)
        for i in range(start_year, end_year + 1):
            x_years_list.append(i)

        self.show_graph(start_year, end_year, x_years_list, y_rp_list, True, "Residential Population for ", "Thousands of Persons")


# State Class Creates a State Object Based on a State's Counties' Data
class State(Geography):

    def __init__(self, name, counties):
        self.name = name
        self.counties = counties

    def __str__(self):
        return self.name

    # Gets Average House Price Index of all Counties
    def get_hpi(self, start_year, end_year):
        hpi_counties_list = []
        for county in self.counties:
            hpi_counties_list.append(county.get_hpi(start_year, end_year))

        hpi_state_list = []
        for i in range(len(hpi_counties_list[0])):
            hpi_state_list.append(0)

        for j in range(len(hpi_counties_list)):
            for i in range(len(hpi_counties_list[0])):
                hpi_state_list[i] += hpi_counties_list[j][i]

        for i in range(len(hpi_state_list)):
            hpi_state_list[i] /= len(self.counties)

        return hpi_state_list


    # Gets Personal Income of all Counties
    def get_pi(self, start_year, end_year):
        pi_counties_list = []
        for county in self.counties:
            pi_counties_list.append(county.get_pi(start_year, end_year))

        pi_state_list = []
        for i in range(len(pi_counties_list[0])):
            pi_state_list.append(0)

        for j in range(len(pi_counties_list)):
            for i in range(len(pi_counties_list[0])):
                pi_state_list[i] += pi_counties_list[j][i]

        for i in range(len(pi_state_list)):
            pi_state_list[i] /= len(self.counties)

        return pi_state_list


    # Gets Total Residential Population of all Counties
    def get_rp(self, start_year, end_year):
        rp_counties_list = []
        for county in self.counties:
            rp_counties_list.append(county.get_rp(start_year, end_year))

        rp_state_list = []
        for i in range(len(rp_counties_list[0])):
            rp_state_list.append(0)

        for j in range(len(rp_counties_list)):
            for i in range(len(rp_counties_list[0])):
                rp_state_list[i] += rp_counties_list[j][i]

        return rp_state_list


# County Class
class County(Geography):
    def __init__(self, name, hpi, pi, rp):
        self.name = name
        self.hpi = hpi
        self.pi = pi
        self.rp = rp

    def __str__(self):
        return self.name

    # Gets Data from Inputed Data Set (Formatted .csv File)
    def get_data(self, start_year, end_year, data_set, column_name):
        data_list = []
        for date, data in zip(data_set["DATE"], data_set[column_name]):
            current_year = int(date[0:4])
            if current_year >= start_year and current_year <= end_year:
                data_list.append(data)
        return data_list

    # Calls get_data() with a County's House Price Index
    def get_hpi(self, start_year, end_year):
        return self.get_data(start_year, end_year, self.hpi, "HPI")

    # Calls get_data() with a County's Personal Income
    def get_pi(self, start_year, end_year):
        return self.get_data(start_year, end_year, self.pi, "PI")

    # Calls get_data() with a County's Residential Population
    def get_rp(self, start_year, end_year):
        return self.get_data(start_year, end_year, self.rp, "RP")
```

```python
# Initializes County and State Objects for the Given Data
def init_data():
    global fresno
    fresno = County("Fresno", hpi_fresno, pi_fresno, rp_fresno)
    global los_angeles
    los_angeles = County("Los Angeles", hpi_los_angeles, pi_los_angeles, rp_los_angeles)
    global orange_county
    orange_county = County("Orange County", hpi_orange_county, pi_orange_county, rp_orange_county)
    global riverside_county
    riverside_county = County("Riverside", hpi_riverside_county, pi_riverside_county, rp_riverside_county)
    global sacramento
    sacramento = County("Sacramento", hpi_sacramento, pi_sacramento, rp_sacramento)
    global san_diego
    san_diego = County("San Diego", hpi_san_diego, pi_san_diego, rp_san_diego)
    global san_francisco
    san_francisco = County("San Francisco", hpi_san_francisco, pi_san_francisco, rp_san_francisco)
    global santa_clara
    santa_clara = County("Santa Clara", hpi_santa_clara, pi_santa_clara, rp_santa_clara)
    global santa_cruz
    santa_cruz = County("Santa Cruz", hpi_santa_cruz, pi_santa_cruz, rp_santa_cruz)
    global counties
    counties = [fresno, los_angeles, orange_county, riverside_county, sacramento, san_diego, san_francisco, santa_clara, santa_cruz]
    global california
    california = State("California", counties)
    global counties_with_state
    counties_with_state = [fresno, los_angeles, orange_county, riverside_county, sacramento, san_diego, san_francisco, santa_clara, santa_cruz, california]
init_data()
```

```python
# USER INTERFACE
import ipywidgets as widgets
from IPython.display import display
import numpy as np

# Welcome Message
print("Welcome! This program is designed to display annual house prices for California and 9 of its counties from 1978-2022.")

# Recursive Display Method
def user_diplay():
    print("Please enter information:\n")

    # County Select
    options_as_strings = [str(option) for option in counties_with_state]
    county_state_choice = widgets.Dropdown(options = counties_with_state, description = "Region:")
    display(county_state_choice)

    # Start/End Year Sliders
    start_year_slider = widgets.IntSlider(min = 1978, value = 1978, max = 2022)
    end_year_slider = widgets.IntSlider(min = 1978, value = 2022, max = 2022)
    print("\nEnter Start Year")
    display(start_year_slider)
    print("Enter End Year")
    display(end_year_slider)

    # Continue Button
    button = widgets.Button(description="Continue")
    output = widgets.Output()

    def on_button_clicked(b):
      with output:
          print(county_state_choice.value.show_hpi_graph(start_year_slider.value, end_year_slider.value))
          print(county_state_choice.value.show_pi_graph(start_year_slider.value, end_year_slider.value))
          print(county_state_choice.value.show_rp_graph(start_year_slider.value, end_year_slider.value))
          print("\n\n\n")
          user_diplay()

    button.on_click(on_button_clicked)
    display(button, output)

user_diplay()
```
